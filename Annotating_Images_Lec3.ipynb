#Annotating Images


In this notebook we will cover how to annotate images using OpenCV. We will learn how to peform the following annotations to images.

Draw lines
Draw circles
Draw rectangles
Add text
These are useful when you want to annotate your results for presentations or show a demo of your application. Annotations can also be useful during development and debugging.

```python
import cv2
import matplotlib.pyplot as plt

from google.colab.patches import cv2_imshow

# Replace the image path with your actual image path
image_path = "/content/drive/MyDrive/Live Data/Live Animals/OIG (11).jpg"
# Read the image using OpenCV
img = cv2.imread(image_path)

# Display the image in the Colab notebook
cv2_imshow(img)
```

```python

import numpy as np

# Draw a rectangle on the image
start_point = (100, 100)  # Top-left corner of the rectangle
end_point = (300, 400)    # Bottom-right corner of the rectangle
color = (0, 25, 0)       # RGB color (green in BGR)
thickness = 2             # Thickness of the rectangle border
img_rectangle = cv2.rectangle(img.copy(), start_point, end_point, color, thickness)

# Add text annotation to the image
text_position = (50, 50)
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1
font_color = (255, 0, 0)  # RGB color (blue in BGR)
font_thickness = 2
text = "Annotated Text"
img_annotated = cv2.putText(img_rectangle.copy(), text, text_position, font, font_scale, font_color, font_thickness)

# Display the original and annotated images
cv2_imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
cv2_imshow(cv2.cvtColor(img_annotated, cv2.COLOR_BGR2RGB))
```

```python

# The line starts from (200,100) and ends at (400,100)
# The color of the line is YELLOW (Recall that OpenCV uses BGR format)
# Thickness of line is 5px
# Linetype is cv2.LINE_AA

cv2.line(img, (200, 200), (400, 100), (0, 255, 255), thickness=9, lineType=cv2.LINE_AA);

# Display the image
plt.imshow(img[:,:,::-1])
```

```python
# Draw a circle on the image
center_coordinates = (200, 100)  # Center of the circle
radius = 80                       # Radius of the circle
color = (0, 255,0)               # RGB color (red in BGR)
thickness = 2                     # Thickness of the circle border
img_with_circle = cv2.circle(img.copy(), center_coordinates, radius, color, thickness)

# Display the original and image with the drawn circle
cv2_imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
cv2_imshow(cv2.cvtColor(img_with_circle, cv2.COLOR_BGR2RGB))
```

```python

# Read the image using OpenCV
img = cv2.imread(image_path)

# Draw a rectangle on the main image
start_point = (100, 100)  # Top-left corner of the rectangle
end_point = (30, 400)    # Bottom-right corner of the rectangle
color = (0, 255, 0)       # RGB color (green in BGR)
thickness = 2             # Thickness of the rectangle border
img_with_rectangle = cv2.rectangle(img.copy(), start_point, end_point, color, thickness)

# Display the original and image with the drawn rectangle
cv2_imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
cv2_imshow(cv2.cvtColor(img_with_rectangle, cv2.COLOR_RGB2BGR))
```

```python
# Add text to the original image
text_position = (50, 80)
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1
font_color = (255, 0, 0)  # RGB color (blue in BGR)
font_thickness = 2
text = "Sample Text"
img_with_text = cv2.putText(img.copy(), text, text_position, font, font_scale, font_color, font_thickness)

# Display the original and image with added text
cv2_imshow(cv2.cvtColor(img_with_text, cv2.COLOR_BGR2RGB))
```

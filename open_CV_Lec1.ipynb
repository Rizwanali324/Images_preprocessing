```python
import cv2
from google.colab.patches import cv2_imshow

# Replace the image path with your actual image path
image_path = "/content/drive/MyDrive/Live Data/Live Animals/OIG (11).jpg"
# Read the image using OpenCV
img = cv2.imread(image_path)

# Display the image in the Colab notebook
cv2_imshow(img)
```

```python

# Split the image into its color channels
b, g, r = cv2.split(img)

# Display each color channel
cv2_imshow(b)
cv2_imshow(g)
cv2_imshow(r)

# Merge the color channels back together
merged_img = cv2.merge([b, g, r])

# Display the original and merged images
cv2_imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
cv2_imshow(cv2.cvtColor(merged_img, cv2.COLOR_BGR2RGB))
```

```python

# Convert the image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Display the original and grayscale images
cv2_imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
cv2_imshow(gray_img)
```

```python
# Convert the image to HSV color space
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# Display the original and HSV images
cv2_imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
cv2_imshow(cv2.cvtColor(hsv_img, cv2.COLOR_HSV2RGB))
```

```python
#modify individual channal
# Split the image into its color channels
b, g, r = cv2.split(img)

# Increase the intensity of the blue channel
b = cv2.add(r, 500)  # Add 50 to all pixel values in the blue channel

# Merge the modified channels back together
modified_img = cv2.merge([b, g, r])

# Display the original and modified images
cv2_imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
cv2_imshow(cv2.cvtColor(modified_img, cv2.COLOR_BGR2RGB))
```

#Basic Image Manipulations
In this notebook we will cover how to perform image transformations including:

Accessing and manipulating images pixels
Image resizing
Cropping
Flipping

```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt

from zipfile import ZipFile
from urllib.request import urlretrieve

from IPython.display import Image

%matplotlib inline
```

```python
import cv2
from google.colab.patches import cv2_imshow

# Replace the image path with your actual image path
image_path = "/content/drive/MyDrive/Live Data/Live Animals/OIG (11).jpg"
# Read the image using OpenCV
img = cv2.imread(image_path)

# Display the image in the Colab notebook
cv2_imshow(img)
```

```python
# prompt: Accessing Individual Pixels

# Access the pixel at location (10, 10)
pixel = img[10, 10]

# Print the pixel value
print(pixel)

# Access all pixels in the image
pixels = img.reshape((-1, 3))

# Print the shape of the pixel array
print(pixels.shape)
```

```python
# prompt: Modifying Image Pixels

# Modify the pixel at location (10, 10) to red
img[1, 1] = (255, 43, 343422)

# Display the modified image
cv2_imshow(img)

# Save the modified image
cv2.imwrite('modified_image.jpg', img)
```

```python
# prompt: Cropping Images

# Crop the image to the region of interest
cropped_image = img[100:200, 100:200]

# Display the cropped image
cv2_imshow(cropped_image)

# Save the cropped image
cv2.imwrite('cropped_image.jpg', cropped_image)
```

```python
# prompt: Crop out the middle region of the image

cropped_image = img[100:200, 100:200]

# Display the cropped image
cv2_imshow(cropped_image)

# Save the cropped image
#cv2.imwrite('cropped_image.jpg', cropped_image)
```

```python
# prompt: resize img

# Resize the image to 50% of its original size
resized_image = cv2.resize(img, (int(img.shape[1] * 0.5), int(img.shape[0] * 0.5)))

# Display the resized image
cv2_imshow(resized_image)

# Save the resized image
#cv2.imwrite('resized_image.jpg', resized_image)
```

```python
# prompt: specify  output size

# Resize the image to 200x200 pixels
resized_image = cv2.resize(img, (200, 440))

# Display the resized image
cv2_imshow(resized_image)

# Save the resized image
#cv2.imwrite('resized_image.jpg', resized_image)
```

```python
# prompt: flip img diff angle and display images in row wise using suplot and  adjust plot size

flipped_image_1 = cv2.flip(img, 0)

flipped_image_2 = cv2.flip(img, 1)

plt.figure(figsize=(10, 10))
plt.subplot(1, 3, 1)
plt.imshow(img)
plt.title('Original Image')
plt.subplot(1, 3, 2)
plt.imshow(flipped_image_1)
plt.title('Flipped Image (Horizontal)')
plt.subplot(1, 3, 3)
plt.imshow(flipped_image_2)
plt.title('Flipped Image (Vertical)')
plt.show()
```
